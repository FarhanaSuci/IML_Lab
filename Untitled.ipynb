```python
import pytesseract
from PIL import Image
import pandas as pd

def extract_text_to_csv(image_path, csv_path):
    # Load the image
    image = Image.open(image_path)
    
    # Extract text using Tesseract OCR
    extracted_text = pytesseract.image_to_string(image, lang="eng+ben")
    
    # Split text into lines
    lines = extracted_text.strip().split("\n")
    
    # Convert extracted text into a structured format (list of lists)
    data = [[line] for line in lines]
    
    # Convert to DataFrame
    df = pd.DataFrame(data, columns=["Extracted Text"])
    
    # Save as CSV file
    df.to_csv(csv_path, index=False)
    
    print(f"Text extracted and saved to {csv_path}")

# Example usage
image_path = "WhatsApp Image 2025-01-21 at 11.03.43 AM.jpeg"  # Replace with the actual image path
csv_path = "extracted_text.csv"  # Replace with the desired CSV file path
extract_text_to_csv(image_path, csv_path)
```
